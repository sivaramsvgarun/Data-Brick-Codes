```python
import pandas as pd
from pyspark.sql import SparkSession
# %pip install xlrd


# Step 1: Read Excel file using pandas
excel_path = "/Volumes/workspace/default/practice-files/Sample - Superstore.xls"
pdf = pd.read_excel(excel_path)

# Step 2: Convert pandas DataFrame to Spark DataFrame
spark = SparkSession.builder.getOrCreate()
df = spark.createDataFrame(pdf)

# Step 3: Write to Parquet format
df.write.mode("overwrite").parquet("/Volumes/workspace/default/practice-files/Sample_Superstore.parquet")
```
